# TESTES PARAMÉTRICOS

## Teste de normalidade

```python
from scipy.stats import normaltest

stat_test, p_valor = normaltest(dados.Renda)
print(stat_test)
print(p_valor)
# Como 0 é < 0.05, então negamos Ho (é normal) e aceitamos H1 (Não é normal).
```

```python
stat_test, p_valor = normaltest(dados.Altura)
print(stat_test)
print(p_valor)
# Como 0  não é < 0.05, então aceitamos Ho (é normal).
```

## Teste Bicaudal (As duas formas realizam a mesma coisa)

```python
from statsmodels.stats.weightstats import ztest

ztest(x1 = dados['Altura'],value = 500 )#É a média que estamos querendo usar como Ho. 
# Padrão é two-sided .alternative :Larger ou smaller
```

## Teste para duas amostras

```python
homens = dados.query('Sexo == 0').sample(n = 500, random_state = 101).Renda
mulheres = dados.query('Sexo == 1').sample(n = 500, random_state = 101).Renda
```

```python
from statsmodels.stats.weightstats import DescrStatsW, CompareMeans

test_H = DescrStatsW(homens)
test_M = DescrStatsW(mulheres)

test_A = test_H.get_compare(test_M)

z, p_valor = test_A.ztest_ind(alternative='larger', value=0)#value = 0 pq a diferença entre o homem a mulher seria 0 (a-b = 0)
p_valor
```

```python
# Ou usando o CompareMeans
test_B = CompareMeans(test_H, test_M)

z, p_valor = test_B.ztest_ind(alternative='larger', value=0)
p_valor
```
