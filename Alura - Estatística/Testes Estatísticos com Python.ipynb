# Testes de uma amostra ( aula 3) aula inutil

```
print("Média dos filmes com pelo menos 10 votos", nota_media_dos_filmes_com_pelo_menos_10_votos.mean()) #parece que a humanidade dá notas altas
```

```
import matplotlib.pyplot as plt # ta querendo mostrar que aumentando o número de elementos a média fica mais precisa
import numpy as np

np.random.seed(75243)
temp = nota_media_dos_filmes_com_pelo_menos_10_votos.sample(frac=1)#tá querendo ver a média de uma amostra cada vez maior

medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias) #Observamos que dps de um certo número de elementos a média fica estável. Mas será que podemos dizer que estes dados representam ou não a média global?
```

```
from statsmodels.stats.weightstats import zconfint #para descobrir a quantidade de elementos e a confiança que temos para a média da amostra

zconfint(nota_media_dos_filmes_com_pelo_menos_10_votos) #retorna o intervalo de confiança pros filmes GLOBAIS ( em geral)
#Os filmes fora da amostra vão estar entre 3,3111. e 3,4528 com 95% de confiança
```

```
#vai usar esse descrstats pra poder ver um intervalo de confiança com base no T student
from statsmodels.stats.weightstats import DescrStatsW #usando o teste t student( o que é uma burrice, pois n é muito maior do que 30)

descr_todos_com_10_votos = DescrStatsW(nota_media_dos_filmes_com_pelo_menos_10_votos)
descr_todos_com_10_votos.tconfint_mean() #a diferença é praticamente despresível (ele fez isso só pra mostrar que t->z quando n>>>>30)
```

```
from statsmodels.stats.weightstats import ztest#vamos verificar se a média do notas1 é igual a media no mundo real

ztest(notas1.rating, value = 3.4320503405352594) # value<0.05. rejeita h0 (SAO DIFERENTES) #nota_media_dos_filmes_com_pelo_menos_10_votos.mean é 3.43205
```

```
np.random.seed(75241) #o objetivo é mostrar que até n pequeno o p value é muito impreciso (no dia a dia é inutil, pois já sabemos disso)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value = 3.4320503405352594)
  return (i,media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), colors='r') # o laranja é o pvalue
```

# Testes para duas amostras ( aula 5)

```
# queremos descobrir a diferença das notas do toystory para todos os filmes
print(ztest(notas1.rating, notas.rating)) # retorna o Z e o pvalue (3.7* 10^-9 MUITO MENOR DO QUE 0.05)-> existe uma diferença significante
zconfint(notas1.rating, notas.rating) # isso mostre a as 'notas1' estão de 0.279~0.5588 mais altas do que as 'notas'
```

```
from scipy.stats import ttest_ind#esse teste vê a diferença dessas duas amostras

ttest_ind(notas.rating, notas1.rating)#mas não ve o invtervalo de confiança
```

```
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toystory = DescrStatsW(notas1.rating)
comparacao = descr_todas_as_notas.get_compare(descr_toystory)#vamos comparar
comparacao.summary()#mostra todos os valores
#pvalor é muito pequeno (descarta a hipotese que sao iguais) e dá o intervalo de confiança # o intervalo é -0.559 e -0.280
```

```
comparacao.summary(use_t=True) # aqui que tá fazendo o t 
```

```
import matplotlib.pyplot as plt

plt.boxplot([notas.rating, notas1.rating], labels=["Todas as notas", "Toy Story"])
plt.title("Distribuição das notas de acordo com filmes")#visualmente vemos que as médias são diferentes
```

```
#nós assumimos que possui uma normalidade, mas precisamos mostrar que é realmente uma dist normal
from scipy.stats import normaltest

_, p = normaltest(notas1.rating)#vamos verificar se as notas do filme 1 vem de uma forma normal # o _ faz com que o primeiro valor que ele retorna seja "pulado"
p #como é <0.05, temos que não podemos considerar que é uma dist normal. PRECISAREMOS USAR TESTES NÃO PARAMÈTRICOS
```
