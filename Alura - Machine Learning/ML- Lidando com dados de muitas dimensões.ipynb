## 4.1 SelectKbest

```python
from sklearn.feature_selection import SelectKBest#para selecionar as melhores "features"(no caso, os melhores exames para nosso estudo)
from sklearn.feature_selection import chi2

selecionar_kmelhores = SelectKBest(chi2, k = 5)#k é o número de features
# "a pessoa precisa fazer só 5 exames e ter a acuracia do nosso modelo"
```

```python
treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v5, 
                                                        diagnostico,
                                                        test_size = 0.3)

selecionar_kmelhores.fit(treino_x,treino_y)#dá um erro pq a função chi2 não aceita valores negativos
#então precisamos voltar lá na v1, pois não havia sido normalizado aida
```

```python
valores_exames_v6 = valores_exames_v1.drop(columns=(["exame_4", "exame_29", "exame_3", "exame_24"]))#excluimos as colunas que vimos serem inuteis para nosso estudo
```

```python
SEED= 1234 #agora rodamos de novo com os valores do exame v6
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6, 
                                                        diagnostico,
                                                        test_size = 0.3)

selecionar_kmelhores.fit(treino_x,treino_y)
treino_kbest = selecionar_kmelhores.transform(treino_x)
teste_kbest = selecionar_kmelhores.transform(teste_x) #encontra as 5 melhores colunas
```

```python
teste_kbest.shape #vemos que só tem 5 colunas, que são nossos exames
```

```python
classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_kbest, treino_y)
print("Resultado da classificação %.2f%%" %(classificador.score(teste_kbest,teste_y)*100))
#o resultado não mudou praticamente nada, testando somente 5 exames
```

## 4.2 Matriz de confusão

```python
from sklearn.metrics import confusion_matrix#reduzimos para 92.4%. Mas será que eu tenho algum outro impacto?
#essa confusionmatrix pega 
matriz_confusao = confusion_matrix(teste_y,classificador.predict(teste_kbest))#vamos usar um heatmap pra visualizar
```

## 4.3 Seleção com RFE

```python
from sklearn.feature_selection import RFE#é uma técnica semlhante ao selectkbest. 
#Ele faz o classificador dar um valor para as features, e vai eliminando a de menor importancia (recurse feature elimination)
SEED= 1234
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6, 
                                                        diagnostico,
                                                        test_size = 0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)#o classificador é o nosso estimador, vai escolher qual feature vai escolher
classificador.fit(treino_x, treino_y)
selecionador_rfe = RFE(estimator = classificador, n_features_to_select = 5, step = 1)#queremos 5 features
selecionador_rfe.fit(treino_x, treino_y)
treino_rfe = selecionador_rfe.transform(treino_x)#vai transformar nosso treino_x e o teste_x
teste_rfe = selecionador_rfe.transform(teste_x)#os resultados serão nosso novo treino e novo teste
classificador.fit(treino_rfe, treino_y)

matriz_confusao = confusion_matrix(teste_y,classificador.predict(teste_rfe))
plt.figure(figsize = (10, 8))
sns.set(font_scale= 2)
sns.heatmap(matriz_confusao, annot = True, fmt = "d").set(xlabel = "Predição", ylabel= "Real")

print("Resultado da classificação %.2f%%" %(classificador.score(teste_rfe,teste_y)*100))
#o resultado foi um pouco pior que o com o selectkbest. eke tá errando mais no cnacer malignos
```

## 5.1 Seleção com RFECV (corss validation)

```python
from sklearn.feature_selection import RFECV#vai dividir os daod em blocos, vai aplicar em cada um, e dps vai encontrar qual é o menor melhor numero
#de features que precisamos para gerar o melhor resultado possível e nos fala quais são
SEED= 1234
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6, 
                                                        diagnostico,
                                                        test_size = 0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_x, treino_y)
selecionador_rfecv = RFECV(estimator = classificador, cv = 5, step = 1, scoring="accuracy")#cv é o numero de divisoes, scoring é sobre oq ele vai avaliar
selecionador_rfecv.fit(treino_x, treino_y)
treino_rfecv = selecionador_rfecv.transform(treino_x)
teste_rfecv = selecionador_rfecv.transform(teste_x)
classificador.fit(treino_rfecv, treino_y)

matriz_confusao = confusion_matrix(teste_y,classificador.predict(teste_rfecv))
plt.figure(figsize = (10, 8))
sns.set(font_scale= 2)
sns.heatmap(matriz_confusao, annot = True, fmt = "d").set(xlabel = "Predição", ylabel= "Real")

print("Resultado da classificação %.2f%%" %(classificador.score(teste_rfecv,teste_y)*100))
#o resultado foi de 94%. melhora o benigno e erra um amais no maligno
```
