```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dados = pd.read_csv('IBGE_alura.csv')
```

```python
dados[:3]
```

```python
x=dados[['Cor','Renda','Altura']]
y=dados['Sexo']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

SEED=12321
np.random.seed(SEED)
raw_x_train,raw_x_test,y_train,y_test = train_test_split(x,y,train_size = 0.75,stratify = y)

escalador = StandardScaler()
escalador.fit(raw_x_train)
x_train=escalador.transform(raw_x_train)
x_test = escalador.transform(raw_x_test)

modelo = DecisionTreeClassifier(max_depth =2)
modelo.fit(x_train,y_train)
previsoes = modelo.predict(x_test)

acuracia = accuracy_score(y_test,previsoes)
print(f'A acuracia é : {(acuracia*100).round(2)} %.')
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier()
dummy.fit(x_train,y_train)
acuracia = dummy.score(x_test,y_test)*100

print(f'A acuracia do dummy é {acuracia.round(2)} %')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
SEED = 12321
np.random.seed(SEED)

cv = StratifiedKFold(n_splits = 10,shuffle = True)
modelo = DecisionTreeClassifier(max_depth=2)
results = cross_validate(modelo,x,y,cv = cv,return_train_score=False)
media = results['test_score'].mean().round(2)
desv_padrao = results['test_score'].std().round(2)
print(f'O intervalo da validação cruzada é [{(media-desv_padrao*2)*100}%,{(media+desv_padrao*2)*100}%]')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

SEED=301
np.random.seed(SEED)

escalador = StandardScaler()
modelo = SVC()

pipeline = Pipeline([('transformaçao',escalador),('estimador',modelo)])
cv = StratifiedKFold(n_splits=10)
results = cross_validate(pipeline,x,y,cv = cv,return_train_score=False)

media = results['test_score'].mean().round(2)
desv_padrao = results['test_score'].std().round(2)
print(f'O intervalo da validação cruzada é [{(media-desv_padrao*2)*100}%,{(media+desv_padrao*2)*100}%]')
```
