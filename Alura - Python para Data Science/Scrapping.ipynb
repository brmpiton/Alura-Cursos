```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup 

url = 'https://alura-site-scraping.herokuapp.com/index.php'
```

```python
response = urlopen(url) #urlopen vai no site e vai trazer
html = response.read() #salvamos nessa variavel
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

# Pesquisando com o beautifulsoup ( find() e findAll() )

```python
soup.find('img') # é o mesmo que soup.img
```

```python
soup.findAll('img')  # é o mesmo que soup('img')
# pega todas as img. Pode colocar: soup.findAll('img', limit = 1)[0]
```

```python
soup.findAll(['h1','h2','h3','h4','h5']) # pesquisando tudo em uma lista
```

```python
soup.findAll('p', {"class": "txt-value"}) # Bem importante. Precisamos montar esse dicionário
```

```python
soup.findAll('p', text = "Belo Horizonte - MG")#encontra todos os items que tem o "belo..."
```

```python
soup.findAll('img', alt="Foto") #pega todas as fotos
```

```python
for item in soup.findAll('img', alt="Foto"):
    print(item.get('src'))#pega só o "código" das fotos
```

```python
soup.findAll('p', class_="txt-value") #se vc procurar a palavra class vai dar problema, pq ela é reservada pelo python
```

```python
soup.findAll(text = True)
```
